```python
# Let's get started importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
plt.style.use('ggplot')
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
df.head()
```

```python
# Investigate the Data
df.info()
df.describe()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['YearBuilt', 'GrLivArea', 'SalePrice', 'LotArea']], figsize=(7,7));
```

```python
# Create a plot that shows the SalesPrice Distribution
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['SalePrice'], bins='auto')
ax.set_title('Sale Price Distribution')
ax.set_xlabel('Sale Price')
ax.set_ylabel('Number of Houses')
ax.axvline(df['SalePrice'].mean(), .1, .9, c='black', ls='-.')
```

```python
# Create a plot that shows the LotArea Distribution
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['LotArea'], bins='auto')
ax.set_title('Lot Area Distribution')
ax.set_xlabel('Lot Size in Square Feet')
ax.set_ylabel('Number of Houses')
# ax.axvline(df['LotArea'].mean(), .1, .9, c='black', ls='-.')
```

```python
# Create a plot that shows the Distribution of the overall house condition
fix, ax = plt.subplots(figsize=(6,4))
ax.hist(df['OverallCond'], bins=8)
ax.set_title('Distribution of House Condition')
ax.set_xlabel('Condition on Scale 1-10')
ax.set_ylabel('Number of Houses')
```

```python
# Create a Box Plot for SalePrice
fig, ax = plt.subplots(figsize=(6,4))
ax.boxplot(df['SalePrice']/1000)
ax.set_ylabel('Sale Price in $1k')
```

```python

price = df.corr()['SalePrice'].sort_values()
fig, ax = plt.subplots(figsize=(10,5))
plt.xticks(rotation=90)
plt.gcf().subplots_adjust(bottom=0.3)
ax.set_title('Sale Price Correlation')
ax.set_xlabel('Attibutes')
ax.set_ylabel('Correlation to Sale Price')
ax.plot(price)
```

```python
# price correlation chart
price
```

```python
# Perform an Exploration of home values by age
fig, ax = plt.subplots(figsize=(6,4))
df['Age'] = df['YrSold'] - df['YearBuilt']
age_plot = df.groupby('Age').SalePrice.mean()/1000
age_plot.plot()
plt.title('Average Sale Price by Home Age')
plt.xlabel('Home Age')
plt.ylabel('Sale Price in $1k')
```
